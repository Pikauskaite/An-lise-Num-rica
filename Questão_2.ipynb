```python
import numpy as np
import matplotlib.pyplot as plt

# Definição da função f(x)
def f(x):
  return x**7 - 7*x**6 + 21*x**5 - 35*x**4 + 35*x**3 - 21*x**2 + 7*x - 1

# Geração de 401 pontos equidistantes no intervalo [1 - 2*10^-8, 1 + 2*10^-8]
x = np.linspace(1 - 2*10**-8, 1 + 2*10**-8, 401)

# Avaliação da função f(x) em cada ponto gerado
f_vect = np.vectorize(f)
y = f_vect(x)

# Plotagem do gráfico da função f(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função f(x)')
plt.show()
```
