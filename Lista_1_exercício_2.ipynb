```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# Definindo a função f(x)
def f(x):
    return x**7 - 7*x**6 + 21*x**5 - 35*x**3 - 21*x**2 + 7*x - 1
```

```python
# Criando uma lista com 401 pontos equidistantes no intervalo [1 − 2 *10^−8, 1 + 2*10^−8]
x_values = np.linspace(1-2*10**(-8), 1+2*10**(-8), 401)
```

```python
# Calculando os valores de f(x) para cada ponto
y_values = [f(x) for x in x_values]
```

```python
# Plotando o gráfico
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função f(x)')
plt.show()
```
